```python
import threading, psutil, os, time, random
from datetime import datetime

sleeptime = random.uniform(0,1)

def one(x):
    for x in words:
        time.sleep(sleeptime)
        print(x,sleeptime)
        

def two(y):
    numbers.reverse()
    print(y,sleeptime)
    
def three(z):
    for z in reversed(ints):
        print(z,sleeptime)
    
    
if __name__=="__main__":
    words = ["hello", "world", "hola", "mundo"]
    numbers = ["one", "two","three", "four", "five", "six", "seven"]
    ints = [1,2,3,4,5,6,7]
    firstthread = threading.Thread(target = one, args=(words,))
    secondthread = threading.Thread(target = two, args=(numbers,))
    thirdthread = threading.Thread(target= three, args=(ints,))
    
    firstthread.start()
    secondthread.start()
    thirdthread.start()
    
    firstthread.join()
    secondthread.join()
    thirdthread.join()
```
